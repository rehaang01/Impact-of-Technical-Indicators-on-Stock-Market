<h1>Predictability/Impact of Technical Indicators on Stock Prices</h1>
<br>

<h2>Indicator No.1</h2>
Simple Moving Average (SMA):
It is calculated by taking the average of (usually) closing prices over a period of days.
It helps to filter out the random short term price fluctuations.
<br>
<h2>Indicator No.2</h2>
Exponential Moving Average (EMA):
It gives more weight to recent prices, making it more valuable tham SMA, by focussing on recent prices.
It is calculated by applying a weighting factor to the most recent price data.
<br>
<h2>Indicator No.3</h2>
Supertrend:
It signals buy or sell based on the price's position relative to a dynamic trendline that measures the markets volatility.
Used to identify the trends in a trending market.
<br>
<h2>Indicator No.4</h2>
Moving Average Convergence Divergence (MACD):
This is a very important indicator as it uses EMA to further give insights.
It is calculated by subtracting the 26-period EMA from the 12-period EMA.
A nine-day EMA of the MACD, called the "signal line".
Both of these lines are plotted and then predictions are made
MACD is used to identify changes in the strength, direction, momentum, and duration of a trend.
<br>
<h2>Indicator No.5</h2>
Bollinger Bands:
Bollinger Bands consist of a middle band (SMA) and two outer bands that are standard deviations away from the middle band.
The bands expand and contract based on market volatility and identify overbought and oversold conditions.
<br>
<h2>Indicator No.6</h2>
Relative Strength Index (RSI):
is a momentum oscillator that measures the speed and change of price movements. It oscillates between 0 and 100, with values above 70
indicating overbought conditions and values below 30 indicating oversold conditions.

Calculating MACD:

```python
def calculate_macd(data):
    ema_12 = data['Close'].ewm(span=12, adjust=False).mean()
    ema_26 = data['Close'].ewm(span=26, adjust=False).mean()
    macd = ema_12 - ema_26
    signal = macd.ewm(span=9, adjust=False).mean()
    return macd, signal
```

Calculating RSI:

```python
def calculate_rsi(data, window=14):
    delta = data['Close'].diff(1)
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)
    avg_gain = gain.rolling(window=window).mean()
    avg_loss = loss.rolling(window=window).mean()
    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    return rsi
```

```python
def calculate_ema(data, span):
    return data['Close'].ewm(span=span, adjust=False).mean()
```

ADDING THE COLUMNS IN COPIED DATA:

```python
#MACD and signal line
macd, signal = calculate_macd(data_copy)
data_copy.loc[:, 'MACD'] = macd
data_copy.loc[:, 'Signal'] = signal
#RSI
data_copy.loc[:, 'RSI'] = calculate_rsi(data_copy)
#EMA
data_copy.loc[:, 'EMA_10'] = calculate_ema(data_copy, span=10)
#short term trends
data_copy.loc[:, 'EMA_25'] = calculate_ema(data_copy, span=25)
#medium trends
```

```python
print(data_copy[['Close', 'MACD', 'Signal', 'RSI', 'EMA_10', 'EMA_25']].tail())
```

```python
def plot_interval(data, start_year, end_year):
    interval_data = data.loc[start_year:end_year]
    plt.figure(figsize=(14, 18))
    #CP
    plt.subplot(5, 1, 1)
    plt.plot(interval_data['Close'], label='Close Price')
    plt.title(f'HDFC Bank Close Price ({start_year}-{end_year})')
    plt.legend()
    
    #EMA-10
    plt.subplot(5, 1, 2)
    plt.plot(interval_data['EMA_10'], label='EMA-10 days', color='red')
    plt.title(f'HDFC Bank EMA (10 days) ({start_year}-{end_year})')
    plt.legend()
    
    #EMA-25
    plt.subplot(5, 1, 3)
    plt.plot(interval_data['EMA_25'], label='EMA-25 days', color='blue')
    plt.title(f'HDFC Bank EMA (25 days) ({start_year}-{end_year})')
    plt.legend()
    
    #MACD and Signal line
    plt.subplot(5, 1, 4)
    plt.plot(interval_data['MACD'], label='MACD', color='red')
    plt.plot(interval_data['Signal'], label='Signal Line', color='blue')
    plt.title(f'MACD and Signal Line ({start_year}-{end_year})')
    plt.legend()

    #RSI
    plt.subplot(5, 1, 5)
    plt.plot(interval_data['RSI'], label='RSI', color='green')
    plt.axhline(70, color='red', linestyle='--')
    plt.axhline(30, color='blue', linestyle='--')
    plt.title(f'RSI ({start_year}-{end_year})')
    plt.legend()

    plt.tight_layout()
    plt.show()
```

WORKING OF THE MACD INDICATOR:
There are two lines MACD and Signal line.
If the macd line crosses above the signal line and the gap between the two lines is plotted
and a histogram indiacting the market will have an upward trend and vice versa for down trends.
So a trade should be taken in upward market if the macd crosses signal above but below the zero level,
only then there will be profit.
Similarly for a decling market, short the share when macd crosses below the signal above zero then there will
be a possibility of profit.
